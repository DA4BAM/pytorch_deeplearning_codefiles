# **텐서와 데이터 로더**

## **1. 텐서 이해하기**

```
import torch
import numpy as np
import pandas as pd
```

### (1) 0-way Tensor

```
scalar = torch.tensor(5)
print(scalar)
print(scalar.shape)
```

### (2) 1-way Tensor

```
vector = torch.tensor([1, 2, 3, 4])
print(vector)
print(vector.shape)
```

### (3) 2-way Tensor

```
matrix = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(matrix)
print(matrix.shape)
```

### (4) 3-way Tensor

```
tensor3d = torch.rand(2, 3, 4)
print(tensor3d)
print(tensor3d.shape)
```

## **2. 텐서 다루기**

### (1) 텐서의 데이터 타입 지정

```
int_tensor = torch.tensor([1, 2, 3], dtype=torch.int64)
float_tensor = torch.tensor([1.0, 2.0, 3.0], dtype=torch.float32)
print(int_tensor)
print(float_tensor)
```

### (2) 텐서 속성 확인

```
print(float_tensor)
print(f'Shape: {float_tensor.shape}')
print(f'Datatype: {float_tensor.dtype}')
print(f'Device: {float_tensor.device}')
```

### (3) 텐서 연산하기

```
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])
print('덧셈:', a + b)
print('곱셈:', a * b)
print('제곱:', a ** 2)
```

### (4) 텐서 집계하기

```
x = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
print('전체 합계:', x.sum())
print('열별 합계:', x.sum(dim=0))
print('행별 합계:', x.sum(dim=1))
```

### (5) 텐서 차원 변경

```
x = torch.tensor([[1, 2], [3, 4], [5, 6]])
print('원래 텐서:', x)
reshaped = x.view(2, 3)
print('변형된 텐서:', reshaped)
```

### (6) 텐서 연결과 분리

```
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6]])

# 텐서 연결하기
concatenated = torch.cat([a, b], dim=0)
print('연결된 텐서:\n', concatenated)
```

```
# 텐서 분리하기
x = torch.tensor([1, 2, 3, 4, 5, 6])
split1, split2 = torch.split(x, 3)
print('분리된 텐서:', split1, split2)
```

### (7) 텐서 내적과 행렬곱

```
# 내적
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
dot = torch.dot(a, b)
print('Dot product:', dot)
```

```
# 행렬곱
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[5, 6], [7, 8]])

result = torch.matmul(A, B)
print(result)
```

## **3. Data Loader**

### (1) 넘파이 배열→텐서 변환

```
path = 'https://bit.ly/advertising_csv'
adv = pd.read_csv(path)
target = 'Sales'
x = adv.drop(target, axis=1)
y = adv.loc[:, target]
x_np = x.values
y_np = y.values
print(type(x_np))
print(x_np.shape)
```

```
# 넘파이 배열 x, y를 텐서로 변환
x_tensor = torch.tensor(x_np, dtype=torch.float32)
y_tensor = torch.tensor(y_np, dtype=torch.float32)
print(x_tensor.shape, y_tensor.shape)
```

### (2) x 텐서, y 텐서 → 텐서 데이터 세트 변환

```
from torch.utils.data import DataLoader, TensorDataset
dataset = TensorDataset(x_tensor, y_tensor)
print(dataset[:3])
print(len(dataset))
```

### (3) 텐서 데이터 세트로 데이터 로더 선언

```
# 데이터 로더 생성
data_loader = DataLoader(dataset, batch_size = 32, shuffle = True)
```

```
for n, batch in enumerate(data_loader):
    x_batch, y_batch = batch
    print(f'배치 {n}, x 배치 크기:{x_batch.shape}')
```

## **4. 디바이스 할당: GPU 혹은 CPU에 할당하기**

### (1) 데이터 준비

```
# 라이브러리 로딩
import torch
from torch import nn
import numpy as np
import pandas as pd

# 데이터 로딩 및 x, y 분할
path = 'https://bit.ly/advertising_csv'
adv = pd.read_csv(path)

target = 'Sales'
x = adv.drop(target, axis=1)
y = adv.loc[:, target]

# numpy로 변환한 다음 텐서로 변환
x_np = x.values
y_np = y.values
x_tensor = torch.tensor(x_np, dtype=torch.float32)
y_tensor = torch.tensor(y_np, dtype=torch.float32)
```

### (2) 디바이스 준비


```
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using {device} device.')
```

### (3) 모델과 데이터세트를 디바이스에 할당

```
# 모델을 디바이스(GPU)에 할당하기
model = nn.Sequential( nn.Linear(3, 1) )
model = model.to(device)
```

```
# 데이터 세트를 디바이스에 할당.
x, y = x_tensor.to(device), y_tensor.to(device)
```

```
# 디바이스 확인
print(next(model.parameters()).device)
print(x.device, y.device)
```
